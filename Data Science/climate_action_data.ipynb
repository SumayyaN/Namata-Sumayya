```python
# Climate action data Analysis
#Objective
# 1. Load and inspect the dataset
# 2. Inspect the data for missing values, incorrect data formats, erroneous entries labeled as 'error', and duplicate records
# 3.Exploratory Data Analysis (EDA)
# 4.find Insights
# 5.Export Clean Dataset
```

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
# Load the dataset into Pandas DataFrame
df = pd.read_csv('climate_action_data.csv')
#intial inspection
df.head(20)
```

```python
# More intial inspection
df.info()

df.dtypes
```

```python
# Descriptive statistics
df.describe()
```

```python
df.shape
```

```python

# List of columns that should be numeric
numeric_columns = ['Soil_Moisture(%)','Soil_pH','Temperature(C)','Humidity(%)','Fertilizer_Recommended(kg/ha)','Irrigation_Recommended(mm)']

# Clean and convert
for col in numeric_columns:
    df[col] = df[col].astype(str).str.replace('%', '').str.strip()  # Remove any stray '%' or spaces
    df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert to float
# Check result
print(df.dtypes)

```

```python
df.isnull().sum()
```

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df.head(5)
```

```python
# Handle missing values
# For numerical columns, we can fill missing values with the mean or median
numeric_cols = ['Soil_Moisture(%)','Soil_pH','Temperature(C)','Humidity(%)','Fertilizer_Recommended(kg/ha)','Irrigation_Recommended(mm)']
for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce') #Convert to numeric, coercing errors to NaN
        # fill NaN values with the median of the column
        df[col] = df[col].fillna(df[col].median())
        
# For categorical columns, we can fill missing values with the mode
categorical_cols = ['Crop_Type']
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mode()[0])  # fill NaN values with the mode of the column
# Check again for missing values
print("Missing values after handling:")
print(df.isnull().sum())
```

```python
# Wrong Data Types

numeric_cols = ['Soil_Moisture(%)','Soil_pH','Temperature(C)','Humidity(%)','Fertilizer_Recommended(kg/ha)','Irrigation_Recommended(mm)']
       

for col in numeric_cols:
    if col in df.columns:
        # Remove extreme outliers (beyond 3 standard deviations)
        mean = df[col].mean()
        std = df[col].std()
        df[col] = np.where((df[col] < mean - 3 * std) | (df[col] > mean + 3 * std), np.nan, df[col])
        df[col] = df[col].fillna(df[col].median())
```

```python

# Check for duplicates
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
# Remove duplicates
df = df.drop_duplicates()
print("Number of rows after removing duplicates:", df.shape[0])
# Check for duplicates dates
print("Number of duplicate dates:", df['Date'].duplicated().sum())
# Remove duplicate dates
df = df.drop_duplicates(subset=['Date'], keep='first')

# Shape
df.shape
```

```python
# Final clean dataset
df.info()
# Check empty values again
print("Missing values after all cleaning:")
print(df.isnull().sum())
```

```python

corr_matrix = df[numeric_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, center=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python

# Set up the figure size
plt.figure(figsize=(18, 12))

# Histogram of 'Soil_Moisture(%)'
plt.subplot(3, 2, 1)
sns.histplot(df['Soil_Moisture(%)'], bins=30, kde=True)
plt.title('Soil_Moisture(%) Distribution')

# Histogram of 'Soil_pH'
plt.subplot(3, 2, 2)
sns.histplot(df['Soil_pH'], bins=30, kde=True)
plt.title('Soil_pH Distribution')

# Histogram of 'Temperature(C)'
plt.subplot(3, 2, 3)
sns.histplot(df['Temperature(C)'], bins=30, kde=True)
plt.title('Temperature(C) Distribution')

# Histogram of 'Humidity(%)'
plt.subplot(3, 2, 4)
sns.histplot(df['Humidity(%)'], bins=30, kde=True)
plt.title('Humidity(%) Distribution')

# Histogram of 'Fertilizer_Recommended(kg/ha)'
plt.subplot(3, 2, 5)
sns.histplot(df['Fertilizer_Recommended(kg/ha)'], bins=30, kde=True)
plt.title('Fertilizer_Recommended(kg/ha) Distribution')

# Histogram of 'Irrigation_Recommended(mm)'
plt.subplot(3, 2, 6)
sns.histplot(df['Irrigation_Recommended(mm)'], bins=30, kde=True)
plt.title('Irrigation_Recommended(mm) Distribution')

# Layout adjustment
plt.tight_layout()
plt.show()
```

```python
#Variables Influencing Fertilizer Recommendations
correlation = df.corr(numeric_only=True)
fert_corr = correlation['Fertilizer_Recommended(kg/ha)'].sort_values(ascending=False)
print(fert_corr)
```

```python
# Crop Type with Highest Average Soil Moisture
highest_moisture_crop = df.groupby('Crop_Type')['Soil_Moisture(%)'].mean().sort_values(ascending=False)
print(highest_moisture_crop)
```

```python
#Irrigation Advice for Crops with Avg Temperature > 30C
# Filter crops with average temperature > 30C
high_temp_df = df[df['Temperature(C)'] > 30]

# Group by Crop and calculate avg soil moisture
moisture_in_high_temp = high_temp_df.groupby('Crop_Type')['Soil_Moisture(%)'].mean()
print("Average soil moisture for crops in >30C temperature:\n", moisture_in_high_temp)

# Recommendations
for crop, moisture in moisture_in_high_temp.items():
    if moisture < 30:
        print(f"..Recommend increased irrigation for {crop} (Avg moisture: {moisture:.2f}%)")
    else:
        print(f"-- No extra irrigation needed for {crop} (Avg moisture: {moisture:.2f}%)")
```
